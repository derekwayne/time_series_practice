# Shampoo Sales Data

## Introduction

Using a dataset for shampoo sales, we would like to develop a model for the time series data that may help us forecast future sales.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
```

```python
# load dataset
shampoo = pd.read_csv('shampoo.txt', header=0, index_col=0, date_parser=lambda x: pd.datetime.strptime('195'+x, '%Y-%m'))
shampoo.info()
sales = shampoo.values
```

It would be best to convert the index to a datetime object.

```python
plt.rcParams["figure.figsize"] = (10,7)

plt.title('Shampoo Sales', fontweight = 'bold', color = 'orange', fontsize = '20')
plt.ylabel('Dollars', fontweight = 'bold', color = 'orange', fontsize = '17')
plt.xlabel('Date', fontweight='bold', color = 'orange', fontsize='17', horizontalalignment='center')
plt.xticks(rotation=45, fontsize='10', horizontalalignment='right')
plt.plot(shampoo)
plt.show()
```

By inspecting the plot we can see that there is an obvious upwards linear trend and no obvious seasonality in the data. We difference once to see if we obtain a stationary time series.

```python
shampoo_diff = shampoo.diff(1)
shampoo_diff = shampoo_diff.dropna()

plt.title('Shampoo Sales', fontweight = 'bold', color = 'orange', fontsize = '20')
plt.ylabel('Dollars', fontweight = 'bold', color = 'orange', fontsize = '17')
plt.xlabel('Date', fontweight='bold', color = 'orange', fontsize='17', horizontalalignment='center')
plt.xticks(rotation=45, fontsize='10', horizontalalignment='right')
plt.plot(shampoo_diff)
plt.show()
```

The differenced plot appears much more like a stationary series. We test for stationarity using the augmented Dickey-Fuller unit root test.

```python
results = adfuller(shampoo_diff['Sales'])
print('The p-value of the test on sales is: ' + str(results[1]))
```

Therefore we conclude that the series is stationary after taking first differences.

```python
plot_acf(shampoo_diff)
plt.show()
```

From the ACF plot above, we can determine that an appropriate model is likely of order 1 since all higher order lag values are not statistically different from zero. Since the series requires differences to attain stationarity we will test an ARIMA(1,1,0), ARIMA(2,1,0), and an ARIMA(1,1,2) model.

```python
ar_mod1 = ARIMA(sales, order=(1, 1, 0))
res = ar_mod1.fit()
print(res.summary())
```

```python
ar_mod2 = ARIMA(sales, order=(2, 1, 0))
res = ar_mod2.fit()
print(res.summary())
```

```python
arma_mod = ARIMA(sales, order=(1, 1, 2))
res = arma_mod.fit()
print(res.summary())
```

The ARIMA(1,1,2) model has the lowest BIC value. Although, the other models are nearly equivalent. Next, we can use this model to forecast 5 months into the future.

```python
res.plot_predict(start=1, end=40)
plt.show()
```
